```python
train['Sex']=train['Sex'].map({'male':1,'female':0})
train['Sex']#将性别用数值编码
```

```python
train['Age']=train['Age'].fillna(train['Age'].mean())#用平均值填充Age列的缺失值
```

```python
#Embarked列用众数填充
train['Embarked']=train['Embarked'].fillna(train['Embarked'].mode()[0])
```

```python
#Embarked列独热编码
train=pd.get_dummies(train,columns=['Embarked'],dtype='int')
```

```python
train
```

```python
train['FamilySize']=train['SibSp']+train['Parch']+1
train
```

```python
train['Survived'].value_counts()#查看类标号计数，看看有没有类不平衡问题
```

```python
features=['Survived','Pclass','Sex','Age','SibSp','Fare','Parch','Embarked_C','Embarked_Q','Embarked_S','FamilySize']
features
```

```python
to_train=train[features]
```

```python
to_train
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
#X_train=to_train[['Pclass','Sex','Age','SibSp','Fare','Embarked']]
#Y_train=to_train['Survived']
```

```python
# 定义数值特征
numeric_features = ['Fare', 'Age']
```

```python
#hold-out决策树模型
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(to_train[['Pclass','Sex','Age','SibSp','Fare','Parch','Embarked_C','Embarked_Q','Embarked_S','FamilySize']], to_train['Survived'], test_size=0.25, random_state=42)
# 标准化数据
scaler = StandardScaler()
X_train[numeric_features] = scaler.fit_transform(X_train[numeric_features])
X_test[numeric_features] = scaler.transform(X_test[numeric_features])

dt_model = DecisionTreeClassifier()
dt_model.fit(X_train, y_train)
y_pred_dt = dt_model.predict(X_test)
accuracy_dt = accuracy_score(y_test, y_pred_dt)
print(f"Decision Tree Accuracy: {accuracy_dt}")
```

```python
# hold-out支持向量机模型
from sklearn.svm import SVC
svm_model = SVC()
svm_model.fit(X_train, y_train)
y_pred_svm = svm_model.predict(X_test)
accuracy_svm = accuracy_score(y_test, y_pred_svm)
print(f"SVM Accuracy: {accuracy_svm}")
```

```python
# hold-out随机森林模型
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier()
rf_model.fit(X_train, y_train)
y_pred_rf = rf_model.predict(X_test)
accuracy_rf = accuracy_score(y_test, y_pred_rf)
print(f"Random Forest Accuracy: {accuracy_rf}")
```

```python
# 使用10-Fold进行交叉验证
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.pipeline import Pipeline
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import AdaBoostClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
models = [
    ('DecisionTree',DecisionTreeClassifier()),
    ('SVM', SVC()),
    ('Random Forest', RandomForestClassifier()),
    ('Naive Bayes', GaussianNB()),
    ("AdaBoost",AdaBoostClassifier()),
    ("LR",LogisticRegression()),
    ('gbc',GradientBoostingClassifier())
]
X=to_train[['Pclass','Sex','Age','SibSp','Fare','Parch','Embarked_C','Embarked_Q','Embarked_S','FamilySize']]
y=to_train['Survived']
X[numeric_features] = scaler.fit_transform(X[numeric_features])

for name, model in models:
    pipeline = Pipeline(steps=[
        ('classifier', model)
    ])
    cv_scores = cross_val_score(pipeline, X, y, cv=10, scoring='accuracy')
    print(f"{name} Cross-validation scores: {cv_scores}")
    print(f"{name} Mean cross-validation accuracy: {cv_scores.mean():.4f}")
    print("-" * 40)
```

```python

from sklearn.model_selection import GridSearchCV
# 定义参数网格
param_grid = {
    'n_estimators': [50, 100, 200],  # 提升树的数量
    'learning_rate': [0.1, 0.05, 0.01],  # 学习率
    'max_depth': [3, 4, 5],  # 树的最大深度
    'min_samples_split': [2, 5, 10],  # 内部节点再划分所需最小样本数
    'min_samples_leaf': [1, 2, 4]  # 叶子节点最少样本数
}

# 创建gb模型
gb_clf = GradientBoostingClassifier(random_state=42)

# 创建GridSearchCV对象
grid_search = GridSearchCV(estimator=gb_clf, param_grid=param_grid, cv=10, scoring='accuracy')

# 执行网格搜索
grid_search.fit(X, y)

# 输出最佳参数和最佳评分
print("Best parameters:", grid_search.best_params_)
print("Best cross-validation score:", grid_search.best_score_)

# 在测试集上评估模型性能
best_model = grid_search.best_estimator_
test_score = best_model.score(X_test, y_test)
print("Test set score with best parameters:", test_score)
```
